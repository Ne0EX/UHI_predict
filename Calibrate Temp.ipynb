```python
import rasterio.merge

path = 'C:/Users/Kittiphong/'

dataset1 = rio.open("D:/Project/LC08_L1TP_129050_20191115_20191115_01_RT.tar/LC08_L1TP_129050_20191115_20191115_01_RT/LC08_L1TP_129050_20191115_20191115_01_RT_B10.TIF")
dataset2 = rio.open("D:/Project/New Map/LC08_L1TP_129051_20191115_20191115_01_RT.tar/LC08_L1TP_129051_20191115_20191115_01_RT/LC08_L1TP_129051_20191115_20191115_01_RT_B10.TIF")      

dest, output_transform=rio.merge.merge([dataset1,dataset2])

with rasterio.open(path+"Edit_Surface_TEMP_129050_20151104.tif") as src:
        out_meta = src.meta.copy()    
out_meta.update({"driver": "GTiff",
                 "height": dest.shape[1],
                 "width": dest.shape[2],
                 "transform": output_transform})
with rasterio.open("B10_2015.tif", "w", **out_meta) as dest1:
        dest1.write(dest.astype(rasterio.float64))
```
